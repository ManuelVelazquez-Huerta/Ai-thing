```python
!ls /content/data
!ls /content/data/processed_data
from tensorflow.keras.preprocessing.image import ImageDataGenerator

IMG_SIZE   = (48, 48)
BATCH_SIZE = 64
DATA_DIR   = "/content/data/processed_data"

datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)

train_gen = datagen.flow_from_directory(
    DATA_DIR,
    target_size=IMG_SIZE,
    color_mode="grayscale",
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="training",
    shuffle=True
)

val_gen = datagen.flow_from_directory(
    DATA_DIR,
    target_size=IMG_SIZE,
    color_mode="grayscale",
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="validation",
    shuffle=False
)

num_classes = len(train_gen.class_indices)
class_names = list(train_gen.class_indices.keys())
print("Classes:", class_names)
```

```python
import matplotlib.pyplot as plt
import numpy as np

images, labels = next(train_gen)
plt.figure(figsize=(10,10))
for i in range(16):
    plt.subplot(4,4,i+1)
    img = images[i].squeeze()
    label_idx = np.argmax(labels[i])
    plt.imshow(img, cmap="gray")
    plt.title(class_names[label_idx])
    plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
from tensorflow.keras import layers, models, optimizers

def build_emotion_cnn(input_shape=(48,48,1), num_classes=7):
    model = models.Sequential()

    # Block 1
    model.add(layers.Conv2D(32, (3,3), padding="same", activation="relu",
                            input_shape=input_shape))
    model.add(layers.MaxPooling2D((2,2)))
    model.add(layers.Dropout(0.25))

    # Block 2
    model.add(layers.Conv2D(64, (3,3), padding="same", activation="relu"))
    model.add(layers.MaxPooling2D((2,2)))
    model.add(layers.Dropout(0.25))

    # Block 3
    model.add(layers.Conv2D(128, (3,3), padding="same", activation="relu"))
    model.add(layers.MaxPooling2D((2,2)))
    model.add(layers.Dropout(0.25))

    # Dense head
    model.add(layers.Flatten())
    model.add(layers.Dense(256, activation="relu"))
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense(num_classes, activation="softmax"))

    return model

input_shape = (IMG_SIZE[0], IMG_SIZE[1], 1)   # grayscale
model = build_emotion_cnn(input_shape=input_shape, num_classes=num_classes)

model.compile(
    loss="categorical_crossentropy",
    optimizer=optimizers.Adam(learning_rate=1e-3),
    metrics=["accuracy"]
)

model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

checkpoint_path = "best_emotion_model.keras"

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=8,
    restore_best_weights=True
)

reduce_lr = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=3,
    min_lr=1e-6,
    verbose=1
)

checkpoint = ModelCheckpoint(
    checkpoint_path,
    monitor="val_loss",
    save_best_only=True,
    verbose=1
)

history = model.fit(
    train_gen,
    epochs=35,              # lower if it takes too long
    validation_data=val_gen,
    callbacks=[early_stop, reduce_lr, checkpoint]
)
```

```python
import matplotlib.pyplot as plt

def plot_history(history):
    hist = history.history
    epochs = range(1, len(hist["loss"]) + 1)

    plt.figure(figsize=(12,4))

    plt.subplot(1,2,1)
    plt.plot(epochs, hist["loss"], label="Train Loss")
    plt.plot(epochs, hist["val_loss"], label="Val Loss")
    plt.title("Loss")
    plt.legend()

    plt.subplot(1,2,2)
    plt.plot(epochs, hist["accuracy"], label="Train Acc")
    plt.plot(epochs, hist["val_accuracy"], label="Val Acc")
    plt.title("Accuracy")
    plt.legend()

    plt.tight_layout()
    plt.show()

plot_history(history)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import itertools

val_gen.reset()
y_prob = model.predict(val_gen)
y_pred = np.argmax(y_prob, axis=1)
y_true = val_gen.classes

print("Classes order:", class_names)
print("\nClassification report:")
print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)

def plot_confusion_matrix(cm, classes, normalize=False):
    if normalize:
        cm = cm.astype("float") / cm.sum(axis=1, keepdims=True)
        title = "Normalized Confusion Matrix"
    else:
        title = "Confusion Matrix"

    plt.figure(figsize=(8,6))
    plt.imshow(cm, interpolation="nearest")
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45, ha="right")
    plt.yticks(tick_marks, classes)

    fmt = ".2f" if normalize else "d"
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 ha="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.tight_layout()
    plt.show()

plot_confusion_matrix(cm, class_names, normalize=False)
plot_confusion_matrix(cm, class_names, normalize=True)
```

```python
val_gen.reset()
images, labels = next(val_gen)
preds = model.predict(images)
pred_labels = np.argmax(preds, axis=1)
true_labels = np.argmax(labels, axis=1)

plt.figure(figsize=(12,8))
for i in range(16):
    if i >= len(images):
        break
    plt.subplot(4,4,i+1)
    img = images[i].squeeze()
    plt.imshow(img, cmap="gray")
    t = class_names[true_labels[i]]
    p = class_names[pred_labels[i]]
    color = "green" if t == p else "red"
    plt.title(f"T: {t}\nP: {p}", color=color, fontsize=9)
    plt.axis("off")
plt.tight_layout()
plt.show()


```

```python
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input
from tensorflow.keras import layers, models, optimizers

```

```python
IMG_SIZE_RGB = (96, 96)
BATCH_SIZE = 64
DATA_DIR = "/content/data/processed_data"

rgb_datagen = ImageDataGenerator(
    validation_split=0.2,
    preprocessing_function=preprocess_input,  # MobileNetV2 preprocessing
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)

train_gen_rgb = rgb_datagen.flow_from_directory(
    DATA_DIR,
    target_size=IMG_SIZE_RGB,
    color_mode="rgb",          #  DIFFERENT: now RGB
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="training",
    shuffle=True
)

val_gen_rgb = rgb_datagen.flow_from_directory(
    DATA_DIR,
    target_size=IMG_SIZE_RGB,
    color_mode="rgb",
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="validation",
    shuffle=False
)

num_classes = len(train_gen_rgb.class_indices)
class_names = list(train_gen_rgb.class_indices.keys())
print("Classes:", class_names)
```

```python
# Build a MobileNetV2-based model
base_model = MobileNetV2(
    input_shape=(IMG_SIZE_RGB[0], IMG_SIZE_RGB[1], 3),
    include_top=False,
    weights="imagenet"
)

# Freeze base model so we only train the new classification head first
base_model.trainable = False

model_mnet = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dropout(0.4),
    layers.Dense(num_classes, activation="softmax")
])

model_mnet.compile(
    optimizer=optimizers.Adam(learning_rate=1e-3),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model_mnet.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

checkpoint_path_mnet = "best_mobilenet_model.keras"

early_stop_mnet = EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True
)

reduce_lr_mnet = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=2,
    min_lr=1e-6,
    verbose=1
)

checkpoint_mnet = ModelCheckpoint(
    checkpoint_path_mnet,
    monitor="val_loss",
    save_best_only=True,
    verbose=1
)

history_mnet = model_mnet.fit(
    train_gen_rgb,
    epochs=20,             #enough
    validation_data=val_gen_rgb,
    callbacks=[early_stop_mnet, reduce_lr_mnet, checkpoint_mnet]
)
```

```python
def plot_history_generic(history, title_suffix=""):
    hist = history.history
    epochs = range(1, len(hist["loss"]) + 1)

    plt.figure(figsize=(12,4))

    plt.subplot(1,2,1)
    plt.plot(epochs, hist["loss"], label="Train Loss")
    plt.plot(epochs, hist["val_loss"], label="Val Loss")
    plt.title(f"Loss {title_suffix}")
    plt.legend()

    plt.subplot(1,2,2)
    plt.plot(epochs, hist["accuracy"], label="Train Acc")
    plt.plot(epochs, hist["val_accuracy"], label="Val Acc")
    plt.title(f"Accuracy {title_suffix}")
    plt.legend()

    plt.tight_layout()
    plt.show()

plot_history_generic(history_mnet, "(MobileNetV2)")
```

```python
val_gen_rgb.reset()
loss_mnet, acc_mnet = model_mnet.evaluate(val_gen_rgb)
print("MobileNetV2 Validation Accuracy: {:.2f}%".format(acc_mnet * 100))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import itertools

val_gen_rgb.reset()
y_prob_mnet = model_mnet.predict(val_gen_rgb)
y_pred_mnet = np.argmax(y_prob_mnet, axis=1)
y_true_mnet = val_gen_rgb.classes

print("Classes order:", class_names)
print("\nMobileNetV2 Classification Report:")
print(classification_report(y_true_mnet, y_pred_mnet, target_names=class_names))

cm_mnet = confusion_matrix(y_true_mnet, y_pred_mnet)

def plot_confusion_matrix(cm, classes, normalize=False, title="Confusion Matrix"):
    if normalize:
        cm = cm.astype("float") / cm.sum(axis=1, keepdims=True)
        title = "Normalized " + title

    plt.figure(figsize=(8,6))
    plt.imshow(cm, interpolation="nearest")
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45, ha="right")
    plt.yticks(tick_marks, classes)

    fmt = ".2f" if normalize else "d"
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 ha="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.tight_layout()
    plt.show()

plot_confusion_matrix(cm_mnet, class_names, normalize=False, title="MobileNetV2 Confusion Matrix")
plot_confusion_matrix(cm_mnet, class_names, normalize=True,  title="MobileNetV2 Confusion Matrix")
```
